*import full charts by weeks date to compare

```python
full_charts = pd.read_csv(r"C:\Users\micha\PycharmProjects\POO_spotify_api\weekly\full_charts.csv")
full_charts.head(10)
```

* filter by first date to add with missing dates

```python
filtered_audio = full_charts.groupby('uri', as_index=False).agg(dates = ('date_week', 'first'))
filtered_audio
```

```python
df_compare = pd.merge(missing_dates, filtered_audio, how='left', on='uri')
df_compare.loc[7, 'release_date'] = '2022'
df_compare
```
