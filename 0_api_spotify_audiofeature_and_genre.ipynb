# API SPOTIFY POUR RECUPERATION GENRES / AUDIO FEATURES / DUREE

```python
import requests
import time
import random
import pandas as pd

# Élaboration des listes de stockage des infos via API
start = 6200


count = start

while start < 6600:
    # Déclaration des listes de stockage en dehors de la boucle
    (ur, img_alb, date_release, dure, son, genr, acous, dans, ener, instru, k, liv, lou, spee, temp, sign, valenc, artist_link) = ([] for i in range(18))

    
    for data in df['uri'][start:start+100]:
        count += 1
        print(f"requête N° {count} :")
        time.sleep(random.randint(10, 20))  # Temporisation entre les appels API

        headers = access_token()
        url_track = f"https://api.spotify.com/v1/tracks/{data}?market=FR"

        try:
            r = requests.get(url_track, headers=headers)
            r.raise_for_status()

            r_json = r.json()
            artist_path = r_json['album']['artists']
            api_artist = artist_path[0]['href']
            api_album = r_json['album']['href']
            image_album = r_json['album']['images'][1]['url']
            release = r_json['album']['release_date']
            duration = r_json['duration_ms']
            sound = r_json['preview_url']
            time.sleep(5)
            r2 = requests.get(api_artist, headers=headers)
            r2.raise_for_status()

            genre = r2.json().get('genres', [])
            print(genre)
            
            time.sleep(5)
            api_audio = f"https://api.spotify.com/v1/audio-features/{data}"
            r3 = requests.get(api_audio, headers=headers)
            r3.raise_for_status()

            content = r3.json()
            acoustic = content['acousticness']
            dance = content['danceability']
            energy = content['energy']
            instrumental = content['instrumentalness']
            key = content['key']
            live = content['liveness']
            loud = content['loudness']
            speech = content['speechiness']
            tempo = content['tempo']
            time_signature = content['time_signature']
            valence = content['valence']

            ur.append(data)
            artist_link.append(api_artist.split('/')[-1])
            img_alb.append(image_album)
            date_release.append(release)
            dure.append(duration)
            son.append(sound)
            genr.append(genre)
            acous.append(acoustic)
            dans.append(dance)
            ener.append(energy)
            instru.append(instrumental)
            k.append(key)
            liv.append(live)
            lou.append(loud)
            spee.append(speech)
            temp.append(tempo)
            sign.append(time_signature)
            valenc.append(valence)

        except requests.exceptions.RequestException as e:
            print(f"An error occurred: {e}")
            ur.append(data)
            artist_link.append(None)
            img_alb.append(None)
            date_release.append(None)
            dure.append(None)
            son.append(None)
            genr.append(None)
            acous.append(None)
            dans.append(None)
            ener.append(None)
            instru.append(None)
            k.append(None)
            liv.append(None)
            lou.append(None)
            spee.append(None)
            temp.append(None)
            sign.append(None)
            valenc.append(None)

        print({
            'uri': data,
            'artist_id': artist_link[-1],
            'image': img_alb[-1],
            'release_date': date_release[-1],
            'duration': dure[-1],
            'sound': son[-1],
            'genre': genr[-1],
            'acoustic': acous[-1],
            'dance': dans[-1],
            'energy': ener[-1],
            'instrumental': instru[-1],
            'key': k[-1],
            'live': liv[-1],
            'loud': lou[-1],
            'speech': spee[-1],
            'tempo': temp[-1],
            'time_signature': sign[-1],
            'valence': valenc[-1],
        })

    df_result = pd.DataFrame({
        'uri': ur,
        'artist_id': artist_link,
        'image': img_alb,
        'release_date': date_release,
        'duration': dure,
        'sound': son,
        'genre': genr,
        'acoustic': acous,
        'dance': dans,
        'energy': ener,
        'instrumental': instru,
        'key': k,
        'live': liv,
        'loud': lou,
        'speech': spee,
        'tempo': temp,
        'time_signature': sign,
        'valence': valenc,
                            })

    df_result.to_csv(f"/datasets/audio_feature_{start}_{start+99}.csv", index=False)
    start += 100  # Incrémenter start pour passer au prochain bloc de 100 URI
    time.sleep(30)  # Temporisation entre les blocs de 100 URI
```
