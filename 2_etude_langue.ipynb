# génération des colonnes langues et détection des colonnes sans chanson

```python
from langdetect import detect
from unidecode import unidecode

# Fonction pour détecter la langue d'un texte donné
def language(chanson):
    if isinstance(chanson, float) and np.isnan(chanson):
        return 'undetermined'
    else:
        return detect(chanson)

# Fonction pour détecter les mauvais lyrics (présence de '/')
def detect_bad_lyrics(words):
    if isinstance(words, float) and np.isnan(words):
        return np.nan
    else:
        for word in words.split():
            if '/' in word:
                return np.nan
        return words

# Fonction pour nettoyer les paroles de chansons en français
def clean_lyrics(lyric, lang):
    if lang == 'fr':
        if isinstance(lyric, float) and np.isnan(lyric):
            return np.nan
        elif len(str(lyric)) > 11:
            clean_chanson = lyric   #or lyric.replace("'", "e ")
            list_of_lines = clean_chanson.split('\n')
            result_chanson = []
            for line in list_of_lines:
                list_of_words = line.split()
                result_line = []
                for word in list_of_words:
                    if "-" in word:
                        split_word = word.replace(',', '').split('-')
                        clean_word = ''.join(split_word[::-1]).lower()
                        result_line.append(unidecode(clean_word))
                    else:
                        result_line.append(unidecode(word.replace(',', '').replace(".", '').replace('(', '').replace(')', '').lower()))
                result_chanson.append(' '.join(result_line))
            full_result_chanson = '\n'.join(result_chanson)
            return full_result_chanson
        else:
            return np.nan
    else:
        return lyric
```

```python
# Apply functions on lyrics to get language and bad scraping rows :
df['lyrics'] = df['lyrics'].apply(detect_bad_lyrics)
df['langue'] = df['lyrics'].apply(language)

# Cleaning lyrics only for french words :
df['lyrics'] = df.apply(lambda row: clean_lyrics(row['lyrics'], row['langue']), axis=1)
```

```python
import re
from collections import defaultdict
```

```python
# function that counts number of sentences :
def num_sentences(text):
    text_replace_punctuations = re.sub(r'[^\w\s]',' ', text.lower())

    unsplited = text_replace_punctuations.splitlines()

    return len(unsplited)

# function that counts the number of total words
def total_words(text):
    text_replace_punctuations = re.sub(r'[^\w\s]',' ', text.lower())
    unsplited = text_replace_punctuations.splitlines()

    count_words = 0
    for word in unsplited:
        count_words += len(word.split())
    return count_words

# function that return total letters without punctuation :
def total_letters(text):
    text_replace_punctuations = re.sub(r'[^\w\s]',' ', text.lower())

    unsplited = text_replace_punctuations.splitlines()
    varchar = ' '.join(unsplited)
    return len(varchar.replace(' ',''))

# return number of unique words :
def unique_words(text):
    text_replace_punctuations = re.sub(r'[^\w\s]',' ', text.lower())

    unsplited = text_replace_punctuations.splitlines()
    varchar = ' '.join(unsplited)
    return len(set(varchar.split()))


# return occurence of unique words :
def occurence(text):
    text_replace_punctuations = re.sub(r'[^\w\s]',' ', text.lower())
    unsplited = text_replace_punctuations.splitlines()
    varchar = ' '.join(unsplited)

    occurrence = defaultdict(int)
    all_listed_words = varchar.split()
    for word in all_listed_words:
        occurrence[word] +=1
    sorted_occurrence = dict(sorted(occurrence.items(), key=lambda item: item[1], reverse=True))
    return dict(sorted_occurrence)




df['num_of_sentences'] = df['lyrics'].apply(num_sentences)
df['total_words'] = df['lyrics'].apply(total_words)
df['total_letters'] = df['lyrics'].apply(total_letters)
df['unique_words'] = df['lyrics'].apply(unique_words)
df['words_occurence'] = df['lyrics'].apply(occurence)
df
```
