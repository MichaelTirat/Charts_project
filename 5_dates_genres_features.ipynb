# ADDING and TRANSFORMING release_date and genre columns

```python
dates = audio['release_date']
list_of_dates = dates.to_list()


small_dates = []
for date in list_of_dates:
    if len(date) < 6:
        small_dates.append(date)
print(len(date))
```

* add a flag to get dates with only years (no month and no day)

```python
audio['dates_flags'] = audio['release_date'].apply(lambda x: 0 if len(x) == 10 else 1)
audio
```

```python
missing_dates = audio[audio['dates_flags']==1]
display(missing_dates.shape)
missing_dates.head()
```

* compare to unique uri with the first entry in charts to get month and day (false but near real situation) : C:\Users\micha\PycharmProjects\Charts_project\datasets\weekly_charts_spotify_full.csv

```python
week_path = r"C:\Users\micha\PycharmProjects\Charts_project\datasets\weekly_charts_spotify_full.csv"
weekly = pd.read_csv(week_path)
weekly.head()
```

```python
weekly_group = weekly.groupby(by='uri', as_index=False).agg(date_week = ('date_week', 'first'))
weekly_group.head()
```

* comparing weekly_group and missing dates

```python
missing_dates_compare = missing_dates.merge(weekly_group,how = 'left', on='uri')
missing_dates_compare.tail()
```

```python
missing_dates_compare.shape

# ok the shape is the same comparing to missing_dates dataframe
```

```python
missing_dates_compare = missing_dates_compare[['uri', 'release_date', 'date_week']]
missing_dates_compare.tail()
```

```python
missing_dates_clean = missing_dates_compare['release_date'].apply(lambda x : x[:4]) + missing_dates_compare['date_week'].apply(lambda x : x[-6:])
missing_dates_clean
```

```python
missing_dates_compare['release_date'] = missing_dates_clean
missing_dates_compare.head()
```

```python
from pprint import pprint
dict_dates = {}
for uri, date in zip(missing_dates_compare['uri'], missing_dates_compare['release_date']):
    dict_dates[uri] = date

pprint(dict_dates)
```

```python
# map sur les uri dont les dates sont incomplètes :
audio['release_date'] = audio.apply(
    lambda row: dict_dates[row['uri']] if row['uri'] in dict_dates else row['release_date'],
    axis=1
)
audio
```

```python
#vérification du mapping :
query = audio.query(f"uri == '02WDuAXXKumemrvsmrVBvH' ")
query
```

```python
# export du dataframe avec les dates clean :
audio.to_csv("datasets/audio_features_clean_dates.csv", index = False, encoding='UTF-8')
```
